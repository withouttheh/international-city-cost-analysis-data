```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
base_url = 'https://www.numbeo.com/cost-of-living/historical-data-city-selector'
```

```python
page = requests.get(base_url)
numbeo_city_soup = BeautifulSoup(page.content, "html.parser")
results = numbeo_city_soup.find('table', class_='related_links')
print(results())
```

```python
list_cities = results.find_all('a')
list_cities
```

```python
city_pages = []
for city in list_cities:
    link_title = city.text
    location = link_title.split(",")
    link_url = city["href"]
    city_obj = {'City': location[0], 
                'Country':location[1], 
                'Url':link_url}
    city_pages.append(city_obj)
```

```python
city_pages[0]
```

```python
df = pd.DataFrame(city_pages)
df.head()
```

```python
selected_cities = ["Cape Town, Nairobi"]
```

```python
for city in selected_cities:
    page = requests.get(df.iloc[df.loc[df['City'] == "Cape Town"].index[0]]["Url"])
    numbeo_one_city_soup = BeautifulSoup(page.content, "html.parser")
    #results = numbeo_city_soup.find('script')
    print(numbeo_one_city_soup)
```
